```python
import matplotlib.pyplot as plt

import numpy as num

y=num.array([80,87,89,56])

mylabels=["Python", "Maths", "English", "Chemistry"]

mycolors=['blue','red','green','yellow']

myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode,colors=mycolors)
plt.show()
```

```python
import matplotlib.pyplot as plt

import numpy as num

y=num.array([30,78,89,90])

mylables=["CSE", "ECE", "Civil", "Mech"]

mycolors=['blue','red','green','yellow']

plt.pie(y,labels=mylables,colors=mycolors)

plt.legend(title="For Subjects")
plt.show()
```
